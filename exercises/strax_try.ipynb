```python
from scipy.optimize import leastsq as ls
from math import *

def drift_time(p,t):
    func=p[0]*e**(-t/p[1])
    return func

def res(p,t,y):
    return drift_time(p,t)-y

p0=[200000,10000]

par,cor,a,b,c=ls(res,p0,args=(coord,peaks), full_output=1)
sq=(cor*np.sum(res(par,coord,peaks)**2)/(len(coord)-len(p0)))[0][0]**0.5
```

```python
plt.scatter(coord,peaks,5)
plt.errorbar(coord,drift_time(par,coord),'r-')
plt.plot(coord,drift_time(par+sq,coord),'g--')
plt.plot(coord,drift_time(par-sq,coord),'g--')
plt.show()
```
